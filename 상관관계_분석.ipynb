### library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 데이터 불러오기

```python
data = pd.read_csv('with_log_concat1.csv', encoding = 'utf-8')
data = data.drop(columns = ['Unnamed: 0'])
```

```python
data
```

## Correlation Matrix

```python
for_corr = data.iloc[:,-4:]
```

```python
for_corr.corr()
```

## Scatter plot

```python
for_corr['log_prod_gdp'] = np.log(data['exp_gdp']*data['imp_gdp'])
```

```python
for_corr.plot.scatter(x = 'log_prod_gdp',y = 'log_Q')
plt.title('Scatter plot between log(exp_gdp*imp_gdp) and log_Q')
plt.show()
```
